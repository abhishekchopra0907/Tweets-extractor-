# Reading tweets from jsonline file 
After writing tweets into the jsonline file, the next step is to extract the required informattion.
we need to extract the following things-
1. text of the tweet
2. date and time of the tweet
3. the number of favorites/likes
4. the number of retweets
5. Number of images present in the tweet

so we need to create a python dictionary before hand with these params to keep a record of information

```python
Table = {
    "Text": [],
    "Created at":[],
    "Number of likes":[],
    "Number of retweets":[], 
    "Number of images":[]
}
```

If we read the JSON File generated then we would understand that there is an attribute called <b>media</b> under the attribute <b>entitties</b>. An attribute named <b>extended_entities</b> is also present if there are multiple media files. so we find the length of list of media to find the  number of images and we push the rest of the data by mapping the params of dictionary to fields of the JSONLines file 

```python
with jsonlines.open('output.jsonl') as reader:
    for obj in reader:
        if 'retweeted_status' in obj and "media" in obj["retweeted_status"]["entities"]:
            if "extended_entities" in obj["retweeted_status"]:
                count = len(obj["retweeted_status"]["extended_entities"])
            else:
                count = 1
        elif "media" in obj["entities"]:
            if "extended_entities" in obj:
                count = len(obj["extended_entities"]["media"])
            else:
                count  = 1
            
        else:
            count = "None"
        
       
        Table["Number of images"].append(count)
        Table["Text"].append(obj["full_text"])
        Table["Created at"].append(obj["created_at"])
        Table["Number of likes"].append(obj["favorite_count"])
        Table["Number of retweets"].append(obj["retweet_count"])
            
    
```

# Converting dictionary to dataframe
we are converting dictionary to dataframe using pandas

```python
df = pd.DataFrame(data = Table)   #converted the dictionary to a dataframe using pandas library
```
